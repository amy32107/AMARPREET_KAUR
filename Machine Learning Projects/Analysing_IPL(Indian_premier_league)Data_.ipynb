```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Load the dataset
ipl_df = pd.read_csv('/content/matches.csv') #Dataset downloaded from kaggle

# Display the first few rows
print(ipl_df.head())
```

```python
# Display basic information about the dataset
print(ipl_df.info())

# Display summary statistics
print(ipl_df.describe())
```

```python
# Check for missing values
print(ipl_df.isnull().sum())

# Drop 'umpire3' if it exists
if 'umpire3' in ipl_df.columns:
    ipl_df = ipl_df.drop(columns=['umpire3'])

# Drop rows with any remaining missing values
ipl_df = ipl_df.dropna()

# Verify that there are no missing values
print(ipl_df.isnull().sum())
```

```python
# Count the number of matches per season
matches_per_season = ipl_df['season'].value_counts().sort_index()

# Plot the number of matches per season
plt.figure(figsize=(10, 6))
sns.barplot(x=matches_per_season.index, y=matches_per_season.values, palette='viridis')
plt.title('Number of Matches per Season')
plt.xlabel('Season')
plt.ylabel('Number of Matches')
plt.xticks(rotation=45)
plt.show()
```

```python
# Count the number of wins per team
wins_per_team = ipl_df['winner'].value_counts()

# Plot the number of wins per team
plt.figure(figsize=(12, 6))
sns.barplot(x=wins_per_team.index, y=wins_per_team.values, palette='coolwarm')
plt.title('Number of Wins by Team')
plt.xlabel('Team')
plt.ylabel('Number of Wins')
plt.xticks(rotation=90)
plt.show()
```

```python
# Count the number of toss decisions
toss_decision_counts = ipl_df['toss_decision'].value_counts()

# Plot the toss decisions
plt.figure(figsize=(8, 6))
sns.barplot(x=toss_decision_counts.index, y=toss_decision_counts.values, palette='Set2')
plt.title('Toss Decisions')
plt.xlabel('Decision')
plt.ylabel('Count')
plt.show()
```

```python
# Count the number of matches per venue
matches_per_venue = ipl_df['venue'].value_counts().head(10)  # Top 10 venues

# Plot the number of matches per venue
plt.figure(figsize=(12, 6))
sns.barplot(x=matches_per_venue.index, y=matches_per_venue.values, palette='magma')
plt.title('Top 10 Venues by Number of Matches')
plt.xlabel('Venue')
plt.ylabel('Number of Matches')
plt.xticks(rotation=90)
plt.show()
```

```python
# Create a new column indicating if the toss winner also won the match
ipl_df['toss_win_match_win'] = ipl_df['toss_winner'] == ipl_df['winner']

# Calculate the percentage of matches where toss winner won the match
toss_win_percentage = ipl_df['toss_win_match_win'].mean() * 100
print(f"Percentage of matches where toss winner won the match: {toss_win_percentage:.2f}%")
```
